### Resphaping, Stacking, Squeezing & Unsqueezing Tesnors
1. Reshaping
    - Reshapes an input tensor to a defined shape
2. View
    - Returns a view of an input tensor, but keeps the original memory of the tensor
3. Stacking
    - Combine multiple tensors of on top of eachother (VStack) or side by side (HStack)
4. Squeeze
    - Removes all `1` dimensions from a tensor
5. Unaqueeze
    - Add a `1` dimension to a target tensor
6. Permute
    - Return a view of the input with dimensions permutated(swapped) in a certain way

```python
x = torch.arange(1,13)
x
```

```python
# Reshape
x_reshaped = x.reshape(3,4)
x_reshaped
```

```python
# View
z = x.view(2,6) # changin z will change x as it is a view
z
```

```python
# Checking if original updates based on view
z[0][0] = 5

z, x
```

```python
# Stack
x1 = torch.arange(0,5)
x2 = torch.arange(0,5)
v_stack = torch.stack([x1,x2], dim=0)
h_stack = torch.stack([x1,x2], dim=1)

print(f"VStack \n{v_stack}\nHStack  \n{h_stack}")
```

```python
# Squeeze & Unsqeueeze
x = torch.tensor([[1,2,3,4,5]])
x.shape
```

```python
x_squeezed = x.squeeze()

x_squeezed, x_squeezed.shape
```

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=1)
x_unsqueezed
```

```python
# Permute
x = torch.rand(224,224,3) # height, width, color channel
x
```

```python
x.permute(2,0,1) # color channel, height, width
```
